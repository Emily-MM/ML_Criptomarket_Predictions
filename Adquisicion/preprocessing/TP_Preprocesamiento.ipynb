# Descentralizados

```python
carpeta_txt = '/content/drive/My Drive/machine/descentralizados'
```

```python
data = []

for archivo in os.listdir(carpeta_txt):
    if archivo.endswith('.txt'):
        ruta_archivo = os.path.join(carpeta_txt, archivo)
        nom = archivo.split('.txt')[0]
        file_data = {}

        with open(ruta_archivo, 'r', encoding='utf-8') as file:
            contenido = file.read()
            secciones = contenido.strip().split("Token: ")[1:]
            for seccion in secciones:
                file_data = {"Exchange": nom}

                file_data["Token"] = seccion.splitlines()[0].strip()
                for line in seccion.splitlines()[1:]:
                    if ':' in line:
                        key, value = line.split(':', 1)
                        key = key.strip()
                        value = value.strip()
                        file_data[key] = value

                data.append(file_data)
```
