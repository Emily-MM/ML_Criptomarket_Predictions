```python
from google.colab import files
import requests
import time

exchanges = [
    "binance", "gdax", "okex", "bybit_spot", "upbit", "kraken", "gate",
    "huobi", "bitfinex", "kucoin", "mxc", "bitget", "crypto_com", "bingx",
    "bitstamp", "bitmart", "bithumb", "lbank", "gemini", "toko_crypto"
]

desired_categories = {
    "Artificial Intelligence", "AI Agents", "Artificial Intelligence (AI)",
    "Real World Assets", "RWA", "Real World Assets (RWA)", "RWA Protocol",
    "Real World Asset Token", "Video Games", "Gaming (GameFi)", "Play to Earn",
    "Gaming Governance Token", "Gaming Utility Token", "Memes", "Meme",
    "Metaverse", "Virtual Worlds", "Governance Tokens", "Utility Tokens",
    "GMCI Meme Index"
}

dont_req = []

def fetch_and_filter_tokens(exchange_id):
    url = f"https://api.coingecko.com/api/v3/exchanges/{exchange_id}/tickers"
    page = 1
    count = 0
    count2 = 0

    filename = f"{exchange_id}.txt"
    with open(filename, 'w') as file:
        while True:
            params = {
                'per_page': 100,
                'page': page
            }
            response = requests.get(url, params=params)
            data = response.json()

            if 'status' in data and data['status'].get('error_code') == 429:
                print("Rate limit exceeded. Waiting...")
                time.sleep(60)
                continue

            if not data.get('tickers'):
                break

            time.sleep(8)
            for ticker in data['tickers']:
                base_coin_id = ticker.get('coin_id')

                if base_coin_id in dont_req:
                    print(f"Skipping {base_coin_id} as it is already in the exclusion list.")
                    continue



                print('request for: ', base_coin_id,'\n')
                base_coin_info = requests.get(f"https://api.coingecko.com/api/v3/coins/{base_coin_id}").json()

                time.sleep(7)

                count += 1
                retries = 1
                while not base_coin_info.get('categories') and retries <= 1:
                    print(f"Retrying for base coin {base_coin_id} due to empty categories...")
                    time.sleep(53)
                    base_coin_info = requests.get(f"https://api.coingecko.com/api/v3/coins/{base_coin_id}").json()
                    retries += 1

                base_categories = base_coin_info.get('categories', [])

                if any(category in desired_categories for category in base_categories):
                    base = ticker.get('base')
                    target = ticker.get('target', 'N/A')
                    market_name = ticker.get('market', {}).get('name', 'Unknown')
                    last_traded_at = ticker.get('last_traded_at')
                    bid_ask_spread_percentage = ticker.get('bid_ask_spread_percentage', 'N/A')
                    volume = ticker.get('volume', 'N/A')

                    base_categories = base_coin_info.get('categories', [])
                    filtered_categories = [category for category in base_categories if category in desired_categories]
                    base_category_str = ', '.join(filtered_categories) if filtered_categories else 'None'

                    base_price_info = base_coin_info.get('market_data', {})
                    base_current_price = base_price_info.get('current_price', {}).get('usd', 'N/A')
                    base_market_cap = base_price_info.get('market_cap', {}).get('usd', 'N/A')
                    base_total_volume = base_price_info.get('total_volume', {}).get('usd', 'N/A')
                    base_high_24h = base_price_info.get('high_24h', {}).get('usd', 'N/A')
                    base_low_24h = base_price_info.get('low_24h', {}).get('usd', 'N/A')
                    base_price_change_24h = base_price_info.get('price_change_24h', 'N/A')
                    base_price_change_percentage_24h = base_price_info.get('price_change_percentage_24h', 'N/A')
                    base_price_change_percentage_7d = base_price_info.get('price_change_percentage_7d', 'N/A')
                    base_price_change_percentage_14d = base_price_info.get('price_change_percentage_14d', 'N/A')
                    base_price_change_percentage_30d = base_price_info.get('price_change_percentage_30d', 'N/A')
                    base_price_change_percentage_60d = base_price_info.get('price_change_percentage_60d', 'N/A')
                    base_price_change_percentage_200d = base_price_info.get('price_change_percentage_200d', 'N/A')
                    base_price_change_percentage_1y = base_price_info.get('price_change_percentage_1y', 'N/A')
                    base_trading_volume_24h = base_price_info.get('total_volume', {}).get('usd', 'N/A')
                    market_cap_change_24h = base_price_info.get('market_cap_change_24h', 'N/A')
                    market_cap_change_percentage_24h = base_price_info.get('market_cap_change_percentage_24h', 'N/A')

                    count2+=1

                    file.write(f"Token: {base}\n")
                    file.write(f"Categories: {base_category_str}\n")
                    file.write(f"Trading Pair: {target}\n")
                    file.write(f"Market: {market_name}\n")
                    file.write(f"Last Traded At: {last_traded_at}\n")
                    file.write(f"Bid/Ask Spread Percentage: {bid_ask_spread_percentage}%\n")
                    file.write(f"Volume: {volume}\n")
                    file.write(f"Current Price (USD): {base_current_price}\n")
                    file.write(f"Market Cap (USD): {base_market_cap}\n")
                    file.write(f"Total Volume (USD): {base_total_volume}\n")
                    file.write(f"High (24h, USD): {base_high_24h}\n")
                    file.write(f"Low (24h, USD): {base_low_24h}\n")
                    file.write(f"Price Change (24h, USD): {base_price_change_24h}\n")
                    file.write(f"Price Change Percentage (24h): {base_price_change_percentage_24h}\n")
                    file.write(f"Price Change Percentage (7d): {base_price_change_percentage_7d}\n")
                    file.write(f"Price Change Percentage (14d): {base_price_change_percentage_14d}\n")
                    file.write(f"Price Change Percentage (30d): {base_price_change_percentage_30d}\n")
                    file.write(f"Price Change Percentage (60d): {base_price_change_percentage_60d}\n")
                    file.write(f"Price Change Percentage (200d): {base_price_change_percentage_200d}\n")
                    file.write(f"Price Change Percentage (1y): {base_price_change_percentage_1y}\n")
                    file.write(f"Trading Volume (24h): {base_trading_volume_24h}\n")
                    file.write(f"Market Cap Change (24h): {market_cap_change_24h}\n")
                    file.write(f"Market Cap Change Percentage (24h): {market_cap_change_percentage_24h}\n")
                    file.write("\n")
                    file.flush()
                    print(count, 'tokens escritos: ', count2, '\n')
                else:
                  dont_req.append(base_coin_id)

            page += 1

fetch_and_filter_tokens("bitfinex")
files.download("bitfinex.txt")
```
